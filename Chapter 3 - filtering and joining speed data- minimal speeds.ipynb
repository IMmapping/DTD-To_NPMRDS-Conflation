```python
import csv
import pandas as pd
import os

path_to_data = os.getcwd() + '//Data//'
#speed_table = path_to_data + 'only-real-time-speeds.csv'
speed_table= path_to_data + 'Here-passenger-car-data.csv'
```

```python
speed_table_df = pd.read_csv(speed_table, low_memory=False)
speed_table_df = speed_table_df.filter(['tmc_code', 'measurement_tstamp', 'speed'])

speed_table_df = speed_table_df[speed_table_df['speed'] > 3]
speed_table_df.head()
```

```python
speed_table_N_df = speed_table_df[speed_table_df['tmc_code'].str.contains('N') == True].append(
                   speed_table_df[speed_table_df['tmc_code'].str.contains('-') == True])

speed_table_N_df.columns = ['N_TMC', 'timestamp', 'speed']

speed_table_P_df = speed_table_df[speed_table_df['tmc_code'].str.contains('N') == False]
speed_table_P_df = speed_table_P_df[speed_table_P_df['tmc_code'].str.contains('-') == False]                

speed_table_P_df.columns = ['P_TMC', 'timestamp', 'speed']

del speed_table_df
```

```python
link_table = (path_to_data + 'dtd_id_to_TMC_link_table.csv')

link_table_df = pd.read_csv(link_table)

matched_N_TMCs = list(set(list([x for x in link_table_df['N_TMC'].values if str(x) != 'nan'])))
matched_P_TMCs = list(set(list([x for x in link_table_df['P_TMC'].values if str(x) != 'nan'])))

del link_table_df

speed_table_P_df = speed_table_P_df[speed_table_P_df['P_TMC'].isin(matched_P_TMCs)]
speed_table_N_df = speed_table_N_df[speed_table_N_df['N_TMC'].isin(matched_N_TMCs)]
```

```python
speed_table_P_df['timestamp'] = pd.to_datetime(speed_table_P_df['timestamp'])
speed_table_P_df['hour'] = speed_table_P_df['timestamp'].dt.hour
speed_table_N_df['minute'] = speed_table_P_df['timestamp'].dt.minute
speed_table_P_df = speed_table_P_df.filter(['P_TMC', 'hour', 'minute','speed'])

speed_table_P_df_4_to_559 = speed_table_P_df[speed_table_P_df['hour'].isin([4,5])]
speed_table_P_df_4_to_559 = speed_table_P_df_4_to_559.filter(['P_TMC', 'speed'])
speed_table_P_df_4_to_559 = speed_table_P_df_4_to_559.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_4_to_559.columns = ['P_TMC','4_559_P']

speed_table_P_df_6_to_659 = speed_table_P_df[speed_table_P_df['hour'] == 6]
speed_table_P_df_6_to_659 = speed_table_P_df_6_to_659.filter(['P_TMC', 'speed'])
speed_table_P_df_6_to_659 = speed_table_P_df_6_to_659.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_6_to_659.columns = ['P_TMC','6_659_P']

speed_table_P_df_7_to_759 = speed_table_P_df[speed_table_P_df['hour'] == 7]
speed_table_P_df_7_to_759 = speed_table_P_df_7_to_759.filter(['P_TMC', 'speed'])
speed_table_P_df_7_to_759 = speed_table_P_df_7_to_759.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_7_to_759.columns = ['P_TMC','7_759_P']

speed_table_P_df_8_to_859 = speed_table_P_df[speed_table_P_df['hour'] == 8]
speed_table_P_df_8_to_859 = speed_table_P_df_8_to_859.filter(['P_TMC', 'speed'])
speed_table_P_df_8_to_859 = speed_table_P_df_8_to_859.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_8_to_859.columns = ['P_TMC','8_859_P']

speed_table_P_df_9_to_959 = speed_table_P_df[speed_table_P_df['hour'] == 9]
speed_table_P_df_9_to_959 = speed_table_P_df_9_to_959.filter(['P_TMC', 'speed'])
speed_table_P_df_9_to_959 = speed_table_P_df_9_to_959.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_9_to_959.columns = ['P_TMC','9_959_P']

speed_table_P_df_10_to_1259 = speed_table_P_df[speed_table_P_df['hour'].isin([10,11,12])]
speed_table_P_df_10_to_1259 = speed_table_P_df_10_to_1259.filter(['P_TMC', 'speed'])
speed_table_P_df_10_to_1259 = speed_table_P_df_10_to_1259.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_10_to_1259.columns = ['P_TMC','10_1259_P']

speed_table_P_df_13_to_1559 = speed_table_P_df[speed_table_P_df['hour'].isin([13,14,15])]
speed_table_P_df_13_to_1559 = speed_table_P_df_13_to_1559.filter(['P_TMC', 'speed'])
speed_table_P_df_13_to_1559 = speed_table_P_df_13_to_1559.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_13_to_1559.columns = ['P_TMC','13_1559_P']

speed_table_P_df_16_to_1859 = speed_table_P_df[speed_table_P_df['hour'].isin([16,17,18])]
speed_table_P_df_16_to_1859 = speed_table_P_df_16_to_1859.filter(['P_TMC', 'speed'])
speed_table_P_df_16_to_1859 = speed_table_P_df_16_to_1859.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_16_to_1859.columns = ['P_TMC','16_1859_P']

speed_table_P_df_19_to_2259 = speed_table_P_df[speed_table_P_df['hour'].isin([19,20,21,22])]
speed_table_P_df_19_to_2259 = speed_table_P_df_19_to_2259.filter(['P_TMC', 'speed'])
speed_table_P_df_19_to_2259 = speed_table_P_df_19_to_2259.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_19_to_2259.columns = ['P_TMC','19_2259_P']

speed_table_P_df_23_to_359 = speed_table_P_df[speed_table_P_df['hour'].isin([23,24,1,2,3])]
speed_table_P_df_23_to_359 = speed_table_P_df_23_to_359.filter(['P_TMC', 'speed'])
speed_table_P_df_23_to_359 = speed_table_P_df_23_to_359.groupby('P_TMC').agg('mean').reset_index()
speed_table_P_df_23_to_359.columns = ['P_TMC','23_359_P']
```

```python
speed_table_N_df['timestamp'] = pd.to_datetime(speed_table_N_df['timestamp'])
speed_table_N_df['hour'] = speed_table_N_df['timestamp'].dt.hour
speed_table_N_df = speed_table_N_df.filter(['N_TMC', 'hour', 'speed'])

speed_table_N_df_4_to_559 = speed_table_N_df[speed_table_N_df['hour'].isin([4,5])]
speed_table_N_df_4_to_559 = speed_table_N_df_4_to_559.filter(['N_TMC', 'speed'])
speed_table_N_df_4_to_559 = speed_table_N_df_4_to_559.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_4_to_559.columns = ['N_TMC','4_559_N']

speed_table_N_df_6_to_659 = speed_table_N_df[speed_table_N_df['hour'] == 6]
speed_table_N_df_6_to_659 = speed_table_N_df_6_to_659.filter(['N_TMC', 'speed'])
speed_table_N_df_6_to_659 = speed_table_N_df_6_to_659.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_6_to_659.columns = ['N_TMC','6_659_N']

speed_table_N_df_7_to_759 = speed_table_N_df[speed_table_N_df['hour'] == 7]
speed_table_N_df_7_to_759 = speed_table_N_df_7_to_759.filter(['N_TMC', 'speed'])
speed_table_N_df_7_to_759 = speed_table_N_df_7_to_759.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_7_to_759.columns = ['N_TMC','7_759_N']

speed_table_N_df_8_to_859 = speed_table_N_df[speed_table_N_df['hour'] == 8]
speed_table_N_df_8_to_859 = speed_table_N_df_8_to_859.filter(['N_TMC', 'speed'])
speed_table_N_df_8_to_859 = speed_table_N_df_8_to_859.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_8_to_859.columns = ['N_TMC','8_859_N']

speed_table_N_df_9_to_959 = speed_table_N_df[speed_table_N_df['hour'] == 9]
speed_table_N_df_9_to_959 = speed_table_N_df_9_to_959.filter(['N_TMC', 'speed'])
speed_table_N_df_9_to_959 = speed_table_N_df_9_to_959.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_9_to_959.columns = ['N_TMC','9_959_N']

speed_table_N_df_10_to_1259 = speed_table_N_df[speed_table_N_df['hour'].isin([10,11,12])]
speed_table_N_df_10_to_1259 = speed_table_N_df_10_to_1259.filter(['N_TMC', 'speed'])
speed_table_N_df_10_to_1259 = speed_table_N_df_10_to_1259.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_10_to_1259.columns = ['N_TMC','10_1259_N']

speed_table_N_df_13_to_1559 = speed_table_N_df[speed_table_N_df['hour'].isin([13,14,15])]
speed_table_N_df_13_to_1559 = speed_table_N_df_13_to_1559.filter(['N_TMC', 'speed'])
speed_table_N_df_13_to_1559 = speed_table_N_df_13_to_1559.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_13_to_1559.columns = ['N_TMC','13_1559_N']

speed_table_N_df_16_to_1859 = speed_table_N_df[speed_table_N_df['hour'].isin([16,17,18])]
speed_table_N_df_16_to_1859 = speed_table_N_df_16_to_1859.filter(['N_TMC', 'speed'])
speed_table_N_df_16_to_1859 = speed_table_N_df_16_to_1859.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_16_to_1859.columns = ['N_TMC','16_1859_N']

speed_table_N_df_19_to_2259 = speed_table_N_df[speed_table_N_df['hour'].isin([19,20,21,22])]
speed_table_N_df_19_to_2259 = speed_table_N_df_19_to_2259.filter(['N_TMC', 'speed'])
speed_table_N_df_19_to_2259 = speed_table_N_df_19_to_2259.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_19_to_2259.columns = ['N_TMC','19_2259_N']

speed_table_N_df_23_to_359 = speed_table_N_df[speed_table_N_df['hour'].isin([23,24,1,2,3])]
speed_table_N_df_23_to_359 = speed_table_N_df_23_to_359.filter(['N_TMC', 'speed'])
speed_table_N_df_23_to_359 = speed_table_N_df_23_to_359.groupby('N_TMC').agg('mean').reset_index()
speed_table_N_df_23_to_359.columns = ['N_TMC','23_359_N']
```

```python
N_speed_table_df = pd.merge(speed_table_N_df_4_to_559, speed_table_N_df_6_to_659)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_7_to_759)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_8_to_859)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_9_to_959)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_10_to_1259)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_13_to_1559)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_16_to_1859)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_19_to_2259)
N_speed_table_df = pd.merge(N_speed_table_df, speed_table_N_df_23_to_359)

os.remove(path_to_data + 'N_TMC_speeds.csv')
N_speed_table_df.to_csv(path_to_data + 'N_TMC_speeds.csv')
```
